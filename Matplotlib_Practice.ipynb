```python
# Import dependencies.
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot using MATLAB Method
plt.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt() Using the Object-Oriented Interface
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot.
plt.plot(x_axis, y_axis)
plt.show()
```

```python
# Create the plot.
plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add a grid.
plt.grid()
# Add the legend.
plt.legend()
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot with ax.plt() Using the Object-Oriented Interface
fig, ax = plt.subplots()
ax.bar(x_axis, y_axis)
```

```python
# Create the plot with ax.plt() Using the Object-Oriented Interface
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis)
```

```python
# Create the plot with ax.plt() switching axes Using the Object-Oriented Interface
fig, ax = plt.subplots()
# change color and add label for key
ax.barh(x_axis, y_axis, color='cyan', label='Chicago')
# Insert legend 
ax.legend()
# Insert Title
ax.set_title("PyBer Fare by Month")
# Label x axis
ax.set_xlabel('Fare($)')
# Label y axis
ax.set_ylabel('Date')
# reverse order the y axis
ax.invert_yaxis()
```
